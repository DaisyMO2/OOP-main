```python
from abc import ABC, abstractmethod
```

```python
class veichle(ABC):
    @abstractmethod
    def display_info(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        
class Car(veichle):
    def display_info(self, make, model, year):
        super().display_info(make, model, year)
        return f"Car Make: {self.make}, Model: {self.model}, Year: {self.year}"
    
    def num_doors(self, doors):
        self.doors = doors
        return f"Number of doors: {self.doors}"
    
O = Car()
print(O.display_info("Toyota", "Camry", 2020))
print(O.num_doors(4))  
```
